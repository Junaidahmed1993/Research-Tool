# Tranformation of the Data Catagories for generation of single data file.

```python
# The cataogories of data in both datasets have same and different columns names
# This is important for the merging the datasets

'''Same column name in both'''
# 1. Authors
# 2. Abstract
# 3. Author Keywords

'''column name with different names'''
# 1. web of science: Article Title: scoups: Title
# 2. web of science: Source Title: scoups: Source title
# 3. web of science: Times Cited, All Databases: scoups: Cited
# 4. web of science: Publication Type: scoups: Document Type
# 5. web of science Publication Year scoups: Year
```

```python
# Now lets harmonized the names of columns which are different

# Harmonize column names for Web of Science
df_wos = df_wos.rename(columns={
    'Article Title': 'Title',
    'Source Title': 'Source title',
    'Times Cited, All Databases': 'Cited',
    'Publication Year': 'Year'
})

# Harmonize column names for Scopus
df_scopus = df_scoups.rename(columns={
    'Authors': 'Authors',
    'Title': 'Title',
    'Year': 'Year',
    'Source title': 'Source title',
    'Cited by': 'Cited',
    'Abstract': 'Abstract',
    'Author Keywords': 'Author Keywords',
    'Document Type': 'Document Type'
})
```

```python
# now lets create a object of column that needed to be retained

columns_to_keep = [
    'Authors', 'Title', 'Source title', 'Year',
    'Cited', 'Abstract', 'Author Keywords', 'Document Type'
]

# Let's merge it

df_wos = df_wos[columns_to_keep]
df_scopus = df_scopus[columns_to_keep]

```

```python
# Let's combine now
df_combined = pd.concat([df_wos, df_scopus], ignore_index=True)
```

# We Should Remove Duplicates:

```python
# So first of all transform the titles

import re

# Function to clean and standardize titles
def preprocess_title(title):
    # Convert to lowercase
    title = title.lower()
    # Remove special characters and keep only alphanumeric and spaces
    title = re.sub(r'[^a-z0-9\s]', '', title)
    # Remove extra spaces
    title = re.sub(r'\s+', ' ', title).strip()
    return title

# Apply preprocessing to the 'Title' column
df_combined['Title'] = df_combined['Title'].apply(preprocess_title)

# Count duplicates after preprocessing
duplicate_count = df_combined.duplicated(subset='Title', keep=False).sum()
print(f"Number of duplicate titles after preprocessing: {duplicate_count}")

# Remove duplicates, keeping the first occurrence
df_combined_unique = df_combined.drop_duplicates(subset='Title', keep='first')
```
